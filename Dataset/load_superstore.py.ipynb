```python
import pandas as pd
from sqlalchemy import create_engine

# Path to your original CSV file
csv_file = r"C:\Users\madha\OneDrive\Desktop\Ecommerce-Sales-Analysis\Dataset\samplesuperstore.csv"

# Read CSV
df = pd.read_csv(csv_file)

print("Rows loaded from CSV:", len(df))
print("Columns:", len(df.columns))

# MySQL connection
engine = create_engine(
    "mysql+pymysql://root:Madhu%4019@localhost/ecommerce_sales"
)

# Upload data
df.to_sql(
    name="superstore",
    con=engine,
    if_exists="replace",
    index=False
)

print("Data uploaded successfully!")
```
